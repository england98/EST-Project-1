```python
for column in df_cleaned.columns:
    if column != 'DateTime' and column != 'Year' and column != 'Month' and column != 'Day' and column != 'Hour':  # 'Date'ì™€ 'Day' ì—´ì€ ê±´ë„ˆë›°ê¸°
        # ê° ì—´ì—ì„œ ê²°ì¸¡ì¹˜ê°€ ìžˆëŠ”ì§€ í™•ì¸
        if df_cleaned[column].isnull().sum() > 0:
            # ì›”ë³„ë¡œ ê·¸ë£¹í™”í•˜ì—¬ í‰ê·  ê°’ ê³„ì‚°
            monthly_means = df_cleaned.groupby(['Year','Month','Day'])[column].transform('mean')
            # ê²°ì¸¡ì¹˜ë¥¼ ì›”ë³„ í‰ê·  ê°’ìœ¼ë¡œ ëŒ€ì²´
            df_cleaned[column].fillna(monthly_means, inplace=True)
```

```python
df_dropped = df_cleaned.drop(columns=['Datetime','Precipitation_mm'])
corr_matrix = df_dropped.corr()

# 2. ížˆíŠ¸ë§µ ê·¸ë¦¬ê¸°
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap')
plt.show()
```

```python
# 1. ë°ì´í„° ì •ê·œí™”
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_dropped)

# 2. PCA ìˆ˜í–‰ (ëª¨ë“  ì£¼ì„±ë¶„ ì‚¬ìš©)
pca = PCA(n_components=None)
pca_result = pca.fit_transform(scaled_data)

# 3. ì„¤ëª…ëœ ë¶„ì‚° ë¹„ìœ¨ í™•ì¸
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 4. ì„¤ëª…ëœ ë¶„ì‚° ë¹„ìœ¨ ê·¸ëž˜í”„ (Elbow Method)
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio)+1), cumulative_variance_ratio, marker='o', label="Cumulative Explained Variance")
plt.title('Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.9, color='r', linestyle='--')  # 90% ì„ 
plt.text(1, 0.92, '90% threshold', color='red')
plt.grid(True)
plt.xticks(range(1, len(explained_variance_ratio)+1))
plt.show()
```
